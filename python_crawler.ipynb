```python
import requests
import os

from bs4 import BeautifulSoup
from PIL import Image
```

```python
url = 'https://www.ghibli.jp/works/karigurashi/#frame'
response = requests.get(url)
soup = BeautifulSoup(response.text)

print(soup.prettify())
```

```python
output_main_dir = 'downloads'

if not os.path.exists(output_main_dir):
    os.makedirs(output_main_dir)
    
work_tags = soup.find_all("a", class_="btn btn-default btn-sm mb4")

work_name = []

for work_tag in work_tags:
    if 'work' not in work_tag['href']:
        continue
    name = work_tag['href'].split('/')[-2]
    if name not in work_name:
        work_name.append(name)
        
work_num = len(work_name)
    
print('有 {} 部作品'.format(work_num))

num = 1

for name in work_name:
    print('------------{name} Start downloading...------------'.format(name=name))
    url = 'https://www.ghibli.jp/works/{name}/#frame'.format(name=name)
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    image_tags = soup.find_all("a", class_="panelarea")

    for img_tag in image_tags:
        if 'gallery' not in img_tag['href']:
            continue

        img_name = img_tag['href'].split('/')[-1]
        img_name = img_name.split('.')[0]
        img_url = 'https://www.ghibli.jp/gallery/{}.jpg'.format(img_name)

        with requests.get(img_url, stream=True) as r:
            r.raise_for_status()
            img = Image.open(r.raw)
            output_dir = output_main_dir + '/{name}'.format(name=name)
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)
            img_save_name = '{outdir}/{img_name}.{img_ext}'.format(outdir=output_dir, img_name=img_name, img_ext=img.format.lower())
            img.save(img_save_name)
        print('Save file {} done.'.format(img_save_name))
    print('------------{} Finish.------------'.format(name))
    num += 1
```
